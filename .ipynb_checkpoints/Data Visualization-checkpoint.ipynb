```python
# Importing all the necessary libraraies.
import numpy as np
import pandas as pd
import joblib, nltk, re, operator
import matplotlib.pyplot as plt
import plotly.express as px

from num2words import num2words
from nltk.util import ngrams
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

from wordcloud import WordCloud, STOPWORDS
from collections import Counter
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
```

## Data Preprocessing.

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def removeLen1words_num2words(text):
    new_text = []
    for word in text.split():
        if word.isnumeric():
            try:
                new_text.append(num2words(word))
            except:
                continue
        else:
            new_text.append(word)

    text = " ".join(new_text)
    text = " ".join([word for word in text.split() if len(word) > 1])
    return text

def PreProcessing(text):
    # Lower-case
    text = str(text).lower()
    
    text = removeLen1words_num2words(text)

    # Remove Stopwords.
    word_tokens = word_tokenize(text)
    text = " ".join([w for w in word_tokens if not w in stop_words])
    
    # Remove urls.
    text = re.sub(r"(http\S+)|(www\S+)|(\w*.com\b)", " ", str(text))

    # Remove Punctuations.
    text = re.sub(r'[+*|/\\\-?.>,<\"\';:!@#$%^&()_`~]', ' ', str(text))
    
    # Lemmetization
    text = " ".join([lemmatizer.lemmatize(word) for word in text.split()])

    # Normalize the text.
    try:
        text = " ".join(normalise(review, variety="BrE", user_abbrevs={}, verbose=False))
    except:
        pass
    
    text = removeLen1words_num2words(text)
    
    return str(text).strip()
```

## 1. WordCloud (Before Preprocessing)

```python
text = " ".join(review for review in df['description'])
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', 
                      max_font_size=1000, max_words=100,
                      colormap='Pastel1', collocations=False, stopwords = STOPWORDS).generate(text)

plt.figure(figsize=(40, 30))
plt.imshow(wordcloud) 
plt.axis("off")
# plt.savefig('./1. wordcloud_before.jpg', bbox_inches='tight')
```

```python
# Preprocessing the description column.
df['preprocessed_description'] = df['description'].apply(PreProcessing)
df
```

## 2. Product Categories Pie Chart.

```python
# Non-Interactive Pie Chart.
categories_counts = dict(df['Primary_Category'].value_counts())
categories_counts_df = pd.DataFrame({'Category':categories_counts.keys(), 'Count':categories_counts.values()})

labels = categories_counts.keys()
sizes = categories_counts.values()
fig, ax = plt.subplots(figsize=(15, 13))
ax.pie(sizes, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax.axis('equal')
# plt.savefig('2. Product Categories Pie Chart.png', bbox_inches='tight')
plt.show()
```

## 6. Top 20 Unigrams distribution.

```python
unigrams = Counter(ngrams((" ".join(df['preprocessed_description'].tolist())).split(), 1))
unigrams = dict(sorted(unigrams.items(), key=operator.itemgetter(1), reverse=True)[:20])
unigrams_df = pd.DataFrame({'unigrams': unigrams.keys(), 'Count': unigrams.values()})
fig = unigrams_df.plot.bar(x='unigrams', y='Count', figsize=(15, 5)).get_figure()
fig.savefig('6. unigrams.jpg', bbox_inches='tight')
```
